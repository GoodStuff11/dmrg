```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

path = r"C:\Users\jonat\OneDrive\Documents\programming\AnacondaProjects\PHYS437B\dmrg\dmrg\outputfile4.csv"
df = pd.read_csv(path)
df["energy noise"] = (df["energy uncertainty"]/df["energy"]).abs()
df["energy diff"] = df["energy"].sub(df.groupby(by=["g", "Nsites"]).transform("min")["energy"])
df["energy diff noise"] = (df["energy diff"]/df["energy"]).abs()
df = df[(df["parity"] == "none") & (df["basis"] == "m") & (df["Nsweep"] == 50) & (df["Nspec"] > 8)]
df = df.sort_values(by=["g", "Nsites", "Nspec"]).reset_index(drop=True)
df["Nspec energy diff"] = df["energy"].diff()
df["Nspec energy diff rel"] = (df["Nspec energy diff"]/df["energy"]).abs()
with pd.option_context('display.max_rows', None, 'display.max_columns', None):  # more options can be specified also
    # display(df.sort_values(by="energy uncertainty"))
    display(df[df['Nspec'] == 10])
```
