```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import uncertainties as u

class ShorthandFormatter(string.Formatter):
    def format_field(self, value, format_spec):
        if isinstance(value, u.UFloat):
            return value.format(format_spec+'S')  # Shorthand option added
        # Special formatting for other types can be added here (floats, etc.)
        else:
            # Usual formatting:
            return super(ShorthandFormatter, self).format_field(value, format_spec)
            
            
def format_number(val, err, opt2=True):
    if opt2:
        frmtr = ShorthandFormatter()
        return frmtr.format("{:u}", u.ufloat(val,err)).replace("(", " (")

    return '{:L}'.format(u.ufloat(val, err)).replace(r"\pm", "&")

frmtr = ShorthandFormatter()


path = r"C:\Users\jonat\OneDrive\Documents\programming\AnacondaProjects\PHYS437B\dmrg\output_data\ed_data\ED_calculations2.csv"
df = pd.read_csv(path)
for g in df['g'].unique():
    print(r"""\begin{table}[h]
\centering
\begin{tabular}{|c c c c c|}
\hline
\# & Energy & SvN & Correlation & Symmetries \\
\hline""")
    tmp = df[df['g']==g]
    for i in range(1,21):
        energy_index = f"E{i}"
        entropy_index = f"SvN{i}"
        correlation_index = f"correlation{i}"
        reflection_symmetry = int(np.isclose(tmp[f"reflection_symmetry{i}"].values[-1],1))
        inversion_symmetry = int(np.isclose(tmp[f"inversion_symmetry{i}"].values[-1],1))
        parity_symmetry = int(np.isclose(tmp[f"parity{i}"].values[-1],1))
        differences = np.abs(tmp[[energy_index,entropy_index, correlation_index]].diff())
        # print(differences)
        # print(differences)
        print(f"{i} & {format_number(tmp[energy_index].values[-1], differences[energy_index].values[-1])} & {format_number(tmp[entropy_index].values[-1], differences[entropy_index].values[-1])} & {format_number(tmp[correlation_index].values[-1], differences[correlation_index].values[-1])} & {(reflection_symmetry, parity_symmetry, inversion_symmetry)} \\\\")

    print("\\hline")
    print("""\\end{tabular}
\\caption{\\(g=%.2f\\)}
\\end{table}""" % g)
    print()
```
