```julia
using Plots

samples = 1
N = 100
threshold = 1.92
probability = []
for i=1:samples
    energy = rand(Float64, N)
    energy = energy ./ (sum(energy)/N)
    bar(energy)
    push!(probability,length(findall(>(threshold), energy))/N)
end

# histogram(probability)
```
